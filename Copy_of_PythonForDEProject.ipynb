```python
import pandas as pd
import matplotlib.pyplot as plt

class SportTurnover:
    def __init__(self):
        self.data = pd.read_csv('sport_turnover_may.csv')
        # Assuming 'TxDateID' is already in the correct datetime format
        self.data['TxDateID'] = pd.to_datetime(self.data['TxDateID'])

    # Cleaning the data
    # Drop all nulls
    def data_clean(self):
        self.data.dropna(inplace=True)

    def total_GGR_per_sport(self):
        return self.data.groupby('SportName')['Total GGR'].sum()

    def top_performing_sport(self):
        return self.data.groupby('SportName')['Total GGR'].sum().idxmax()

    def top_10_GGR_per_sport(self):
        return self.total_GGR_per_sport().nlargest(10)

    def average_daily_trend(self):
        return self.data['Total GGR'].mean()

    def plot_sports_trend(self):
        self.data.groupby('TxDateID')['Total GGR'].sum().plot(kind='line')
        plt.title('GGR over Time')
        plt.xlabel('Date')
        plt.ylabel('Total GGR')
        plt.show()

    def plot_GGR_per_sport(self):
        self.top_10_GGR_per_sport().plot(kind='bar')
        plt.title('GGR per Sport')
        plt.xlabel('Sport')
        plt.ylabel('Total GGR')
        plt.show()

```

```python

turnover = SportTurnover()
#turnover.data_clean()
print('Total GGR per sport:\n', turnover.total_GGR_per_sport())
print('\nTop performing sport: ', turnover.top_performing_sport())
print('\nAverage Daily GGR: ', turnover.average_daily_trend())
turnover.plot_sports_trend()
turnover.plot_GGR_per_sport()
```
